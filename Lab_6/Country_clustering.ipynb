```python
#Kmeans 

from sklearn import cluster

km=cluster.KMeans(n_clusters=3)
km.fit(data.loc[:,['child_mort','life_expec','gdpp','health','income','inflation']])
km.labels_
```

```python
data['clusters'] = km.labels_
```

```python
data[data['clusters'] == 0]
```

```python
data[data['clusters'] == 1]
```

```python
data[data['clusters'] == 2]
```

For this task, K-means seems to be a good choice. I applied K means clustring with 3 clusters. However, the results are not in the shape of 'underdeveloped', 'developing', 'developed'. Instead the results clustered countries as 'underdeveloped - developing', 'developed' and 'Super Rich Countries'

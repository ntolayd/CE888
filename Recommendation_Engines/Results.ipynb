```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
models = ["SVD", "RBM", "Single Layer Autoencoder No Reg", "3 Layer Autoencoder No Reg", "5 Layer Autoencoder No Reg",
          "Single Layer Autoencoder with Reg","3 Layer Autoencoder with Reg","5 Layer Autoencoder with Reg",
          "Shallow MLP","Deeper MLP","NueMF"]


RMSE = [0.942, 2.710, 1.052, 1.154, 0.902, 0.985, 0.932,  0.957, 1.429, 0.890, 0.901]
```

```
df = pd.DataFrame([models, RMSE]).T
df.columns = ["Model","RMSE"]
df = df.sort_values(by="RMSE")
```

```
def annotater(graph):
    for p in graph.patches:
        graph.annotate('{:.1f}'.format(p.get_height()), 
                       (p.get_x() + p.get_width() / 2, p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       textcoords = 'offset points', fontsize=12)

fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x="Model", y="RMSE", data=df, palette=sns.color_palette("muted"))
ax.set_xlabel("Model", fontsize=15)
ax.set_ylabel("RMSE", fontsize=15)
plt.xticks(ticks= list(range(len(df["Model"])+1)), labels = df["Model"], ha='right', fontsize=15, rotation=30, rotation_mode='anchor')
annotater(ax)
plt.show()
```

```
map = [0.0024, 0.0719, 0.0022, 0.0032, 0.0025, 0.0004, 0.0007, 0.0002, 0.0083, 0.0165, 0.0091]
```

```
df_map = pd.DataFrame([models, map]).T
df_map.columns = ["Model","MAP"]
df_map = df_map.sort_values(by="MAP")
```

```
def annotater(graph):
    for p in graph.patches:
        graph.annotate('{:.4f}'.format(p.get_height()), 
                       (p.get_x() + p.get_width() / 2, p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       textcoords = 'offset points', fontsize=12)

fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x="Model", y="MAP", data=df_map, palette=sns.color_palette("muted"))
ax.set_xlabel("Model", fontsize=15)
ax.set_ylabel("MAP@30", fontsize=15)
plt.xticks(ticks= list(range(len(df_map["Model"]))), labels = df_map["Model"], ha='right', fontsize=15, rotation=30, rotation_mode='anchor')
annotater(ax)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('bmh')
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
horflip_widthshift_brightness_rotation = pd.DataFrame({'Xception':70, 'DenseNet121':93, 'VGG16':79, 'ResNet51':90}, index=[0])
rotation_randomnoise = pd.DataFrame({'Xception':63, 'DenseNet121':78, 'VGG16':78, 'ResNet51':42},  index=[0])
rotation_zoom = pd.DataFrame({'Xception':72, 'DenseNet121':76, 'VGG16':84, 'ResNet51':81},  index=[0])
no_aug = pd.DataFrame({'Xception':67, 'DenseNet121':84, 'VGG16':77, 'ResNet51':85},  index=[0])
rotation_brightness = pd.DataFrame({'Xception':74, 'DenseNet121':90, 'VGG16':81, 'ResNet51':80},  index=[0])
rotation_height_width_shift = pd.DataFrame({'Xception':68, 'DenseNet121':73, 'VGG16':78, 'ResNet51':75},  index=[0])
rotation_verticalflip = pd.DataFrame({'Xception':81, 'DenseNet121':93, 'VGG16':72, 'ResNet51':85},  index=[0])
```

```python
Xception = pd.DataFrame({'No_Augmentation':67,'Horizontal_Flip':70, 'Rotation_RandomNoise':63, 
                         'Rotation_Zoom':72, 
                         'Rotation_Brightness':74, 'Rotation_HeightWidth_Shift':68,
                         'Rotation_VerticalFlip':81}, index=[0])

DenseNet = pd.DataFrame({'No_Augmentation':84,'Horizontal_Flip':93, 'Rotation_RandomNoise':78, 
                         'Rotation_Zoom':76, 
                         'Rotation_Brightness':90, 'Rotation_HeightWidth_Shift':73,
                         'Rotation_VerticalFlip':93}, index=[0])

VGG = pd.DataFrame({'No_Augmentation':77,'Horizontal_Flip':79, 'Rotation_RandomNoise':78, 
                         'Rotation_Zoom':84, 
                         'Rotation_Brightness':81, 'Rotation_HeightWidth_Shift':78,
                         'Rotation_VerticalFlip':72}, index=[0])


ResNet = pd.DataFrame({'No_Augmentation':85,'Horizontal_Flip':90, 'Rotation_RandomNoise':42, 
                         'Rotation_Zoom':81, 
                         'Rotation_Brightness':80, 'Rotation_HeightWidth_Shift':75,
                         'Rotation_VerticalFlip':85}, index=[0])

Xception = pd.melt(Xception)
Xception.columns = ['Augmentation','Accuracy']

DenseNet = pd.melt(DenseNet)
DenseNet.columns = ['Augmentation','Accuracy']

VGG = pd.melt(VGG)
VGG.columns = ['Augmentation','Accuracy']

ResNet = pd.melt(ResNet)
ResNet.columns = ['Augmentation','Accuracy']
```

```python
plt.figure(figsize=(10,7))
splot = sns.barplot(x='Augmentation', y='Accuracy', data=Xception)
plt.xticks(rotation=80)
plt.tick_params(axis='both', labelsize=18)
plt.title('Xception')
for p in splot.patches:
    splot.annotate(format(p.get_height()), 
                   (p.get_x() + p.get_width() / 2, p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points',size=15)
```
